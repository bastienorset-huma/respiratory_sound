```python
import os

import pandas as pd
import numpy as np
```

```python
df_info = pd.read_csv('../../data/info_dataset.csv',index_col=0)
```

```python
df_info['location'] = df_info.filename.apply(lambda x: x.split('_')[2])
df_info['sc'] = df_info.filename.apply(lambda x: x.split('_')[3])
df_info['idk'] = df_info.filename.apply(lambda x: x.split('_')[1])
```

```python
df_info['wheeze_b'] = np.where(df_info['wheeze']>0,1,0)
df_info['crackle_b'] = np.where(df_info['crackle']>0,1,0)
df_info['both'] = np.where(df_info[['wheeze_b','crackle_b']].sum(axis=1)>0,1,0)
df_info['none'] = np.where(df_info[['wheeze_b','crackle_b']].sum(axis=1)==0,1,0)
```
